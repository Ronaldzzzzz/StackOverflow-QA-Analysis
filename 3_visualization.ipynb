```python
import pandas, datetime 
import matplotlib.pyplot as plt

Question = pandas.read_sql('Questions', 'sqlite:///QAT_proposed.db')

hist = Question[['CreateTime_hour','FirstAnsTime']].groupby('CreateTime_hour').max()

def time_to_years(x):
    one_year = 24*60*60*365
    return float(x['FirstAnsTime']/one_year)

hist = hist.apply(time_to_years, axis=1)
ax = hist.plot()
plt.xlabel("24H\nFig. 5")
plt.ylabel("Maximum time of answer (years)")
plt.title("Maximum time of answer collected by time")
plt.show()
```

```python
import pandas, datetime 
import matplotlib.pyplot as plt

Question = pandas.read_sql('Questions', 'sqlite:///QAT_proposed.db')

week_index = ['Sun','Mon','Tus','Wed','Thr','Fri','Sat']
data_in_week = []
one_hour = 60*60.0
for weekday in week_index:
    data_in_week.append(Question[Question['CreateTime_weekday'] == weekday]['FirstAnsTime']/one_hour)

bp = plt.boxplot(data_in_week,showfliers=False)
plt.xticks(list(range(1,8)), week_index)
plt.ylabel('hours')
plt.xlabel('weekday\nFig. 11')
plt.title('boxplt of first response time by weekday bins')
plt.savefig('bp.png',dpi=600)
plt.show()
```

```python
import matplotlib.pyplot as plt

Question = pandas.read_sql('Questions', 'sqlite:///QAT_proposed.db')

hour_index = list(range(0,24))
data_in_hour = []
one_hour = 60*60.0
for hour in hour_index:
    data_in_hour.append(Question[Question['CreateTime_hour'] == hour]['FirstAnsTime']/one_hour)

bp = plt.boxplot(data_in_hour,showfliers=False)
plt.xticks(list(range(1,25)), hour_index)
plt.ylabel('response time (hours)')
plt.xlabel('hour\nFig. 12')
plt.title('boxplt of first response time by hour bins')
plt.savefig('bp.png',dpi=600)
plt.show()
```
